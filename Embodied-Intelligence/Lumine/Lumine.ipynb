## 研究背景与目标
1. 目标：构建能在 开放世界（open worlds） 中像人类一样 感知、推理、行动 的通用智能体，是实现 AGI（人工通用智能） 的核心目标之一
2. 现有进展与局限
- **传统智能体**：在封闭环境中表现出色，但是***(1)依赖明确奖励函数。(2)泛化能力差。(3)难以处理开放世界的模糊性与多样性***
- **实现开放世界通用智能体六大挑战**：
- （1）可扩展的环境：环境需丰富、多样、可复现，支持组合性动态
- （2）多模态感知：需融合视觉、GUI、文本等多模态信息
- （3）高层规划：能生成并调整长时程、自激励的计划
- （4）低层控制：将抽象意图转化为精确、可执行的动作
- （5）记忆：维护多时间尺度的经验，支持一致性决策
- （6）实时推理：在严格延迟约束下平衡计算与响应速度
- **新路径：语言赋能的智能体**：
- 自然语言作为通用接口，能桥接多模态感知与复杂决策
- 大语言模型（LLMs）具备常识与推理能力，已开始应用于机器人、游戏等交互环境
- ***但仍然存在短板***：（1）低层控制能力弱。（2）推理效率低，难以满足实时交互需求
