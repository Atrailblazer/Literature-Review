# Bip3D: 强化3D感知的多模态融合框架
### 研究背景与动机
1. 3D感知在实体智能中的重要性
* 3D 感知是实体智能（如机器人、自动驾驶）的关键组成部分，用于理解环境中的物体位置、形状、类别等
* 传统方法主要依赖点云（如 PointNet++、3D CNN），虽然几何信息精确，但存在：<br>
    - 数据稀疏、缺乏纹理
    - 对传感器质量依赖高（如透明、反光物体难以处理）
    - 数据标注成本高、规模有限
2. 2D 视觉基础模型的优势
* 2D 图像数据丰富，预训练模型（如 CLIP、DINO、EVA）具有强大的语义理解和泛化能力
* 本文提出：能否利用 2D 视觉基础模型来提升 3D 感知性能？
### 核心贡献
* BIP3D 是一个以图像为中心的 3D 感知模型，基于 GroundingDINO 架构，支持多视图图像、文本指令和深度图输入，输出 3D 边界框。主要改进包括：
1. 相机建模
- 显式构建相机模型，支持内参和外参输入
- 为 2D 图像特征提供 3D 位置编码，增强几何感知
2. 多视图融合
- 将 2D 可变形注意力改为 3D 形式
- 支持动态多视图特征融合，提升跨视角一致性
3. 多模态融合
- 增加深度图编码分支
- 现图像与深度图的多模态特征融合，提升定位精度
### 模型架构与关键技术
1. 整体流程
- **输入**：多视图 RGB 图像、相机参数、深度图、文本指令
- **编码器**：图像编码器（Swin-Transformer）、文本编码器（BERT），深度编码器（轻量化 ResNet34）
- **特征增强器**：融合文本与图像特征
- **空间增强器**：进行 3D 位置编码与深度融合
- **解码器**：基于 Transformer 的检测头，输出 3D 边界框
2. 关键技术说明
- **空间增强器**：<br>
    - 利用相机参数将 2D 图像特征映射到 3D 空间<br>
    - 预测深度分布，加权生成图像位置嵌入
    - 融合图像、深度与位置信息，增强 3D 感知能力
- **损失函数设计**：<br>
    - 使用一对一的匹配损失（来自 DETR）
    - 包含分类损失、中心点回归损失、边界框回归损失
    - 边界框回归使用 Wasserstein 距离，避免方向歧义
