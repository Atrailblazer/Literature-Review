# 自适应大邻域搜索算法
## 论文
1. An adaptive large neighborhood search method for the drone–truck arc routing problem(无人机-货车圆弧路径问题的自适应大邻域搜索方法)(SCI一区)
2. A multi-visit flexible-docking vehicle routing problem with drones for simultaneous pickup and delivery services
3. Optimizing bidirectional delivery with multiple drones and trucks

# 强化学习/深度学习
## 论文
1. Dynamic collaborative truck-drone delivery with en-route synchronization and random requests
2. An Unmanned Delivery Vehicle Path-Planning Method Based on Point-Graph Joint Embedding and Dual Decoders
- 提出基于点图联合嵌入与双解码器（PGDD）的深度强化学习模型，结合注意力机制与伪标签学习，解决了无人配送车在模糊时间窗和客户满意度约束下的大规模路径规划问题
3. Deep Reinforcement Learning for Trajectory Path Planning and Distributed Inference in Resource-Constrained UAV Swarms
- 提出一种基于深度强化学习（PPO算法）的联合轨迹规划与分布式CNN推理优化模型，在资源受限的无人机群中，实现低延迟的图像分类协同推理，同时优化飞行路径以降低干扰与传输延迟
4. Hierarchical task network-enhanced multi-agent reinforcement learning: Toward efficient cooperative strategies
- 提出HS-MARL框架，将分层任务网络（HTN）与符号选项结合到多智能体强化学习中，引入内部奖励机制引导探索，在稀疏奖励和次优条件下，提升多智能体协作任务的探索效率与策略稳定性。
5. MAGNNET: Multi-Agent Graph Neural Network-based Efficient Task Allocation for Autonomous Vehicles with Deep Reinforcement Learning
- 提出一种基于图神经网络与多智能体深度强化学习的CTDE框架，解决了异构多智能体在通信受限环境下的动态任务分配与冲突避免问题
6. Sequential Stochastic Combinatorial Optimization Using Hierarchical Reinforcement Learning（ICLR 2025）
- 提出了一种基于“醒-睡”训练程序的双层选项框架（WS-option），使用分层强化学习方法，解决了序列随机组合优化问题，如自适应影响力最大化和路径规划，能够同时优化预算分配与节点选择。
7. Hierarchical RNNs with graph policy and attention for drone swarm（计算设计与工程领域的专业期刊，质量中等）
- 提出了一种结合分层RNN、图策略与注意力机制的方法，解决了无人机群的编队控制与路径规划问题，实现了多智能体在时空依赖下的协同决策与障碍物规避

# 模拟退火算法
1. The multi-visit drone routing problem for pickup and delivery services
- 阶段启发式算法，结合最大有效载荷方法构建初始解，并采用改进的模拟退火算法，融入问题特定的邻域算子和加速策略进行优化，以求解多访问无人机取送路径规划问题
2. Collaborative truck multi-drone routing and scheduling problem: Package delivery with flexible launch and recovery sites
- 构建混合整数线性规划模型并设计模拟退火与变邻域搜索相结合的两阶段启发式算法，解决了卡车与多异构无人机协同配送中允许在非客户位置起降的路径规划与调度优化问题
